```python
# Encode and run query to get match data for The International
query = """
SELECT 
    m.*,
    l.name as league_name,
    l.leagueid
FROM matches m
JOIN leagues l USING(leagueid)
WHERE l.name ILIKE 'The International%'
ORDER BY m.start_time DESC
"""

encoded_query = data_extraction.encode_query(query)
base_url = "https://api.opendota.com/api/explorer"
params = {"sql": encoded_query}

match_data = data_extraction.get_api_data(base_url, params=params)
```
